# Spectrum cycler — AGN vs galaxy visual check

Click **◀ / ▶** to cycle through the KEEP spectra one at a time, toggle between
**Galaxy** and **AGN** with the radio buttons. NaN spectra (galaxies outside the VDFI
extraction footprint) are skipped automatically. Each view shows the observed-frame
and rest-frame (near Lyα) spectrum, with the galaxy's ID, z, and AGN status.

Uses the product you read with `core.read_core_fits(...)` plus the KEEP catalog (with
`AGN_FLAG` / `CLEAN`), matched by the `ID` column.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
from astropy.table import Table, vstack
from utils_lya_halo import core, stack

# ---- load the product and the KEEP catalog ----------------------------
catalog_keep = Table.read('catalogs/lya_halo_catalog_cuts/lya_mosdef_catalog_KEEP.fits', format='fits')

prod_egs    = core.read_core_fits("outputs_core/6-23-26-core_extraction_egs_MF.fits")
prod_cosmos = core.read_core_fits("outputs_core/6-23-26-core_extraction_cosmos_MF.fits")
prod = stack.concat_core_prods(prod_egs, prod_cosmos)

LYA_REST = 1215.67
REST_HALF_WIN = 100.0

catalog_keep
```

## 1. Build AGN / galaxy index lists (NaN-skipped)

Match the extraction product to the KEEP catalog by `ID`, tag each product row as AGN or
clean galaxy, and drop rows whose spectrum is all-NaN (outside the VDFI footprint).

```python
prod_cat = prod["catalog"]
spec_all = np.asarray(prod["core_spec"], float)
n_prod   = len(prod_cat)

# map ID -> AGN flag from the KEEP catalog
keep_id   = np.asarray(catalog_keep["ID"]).astype(str)
keep_agn   = catalog_keep["AGN_FLAG"]
keep_clean = catalog_keep["CLEAN"]
agn_of  = {i: a for i, a in zip(keep_id, keep_agn)}
clean_of= {i: c for i, c in zip(keep_id, keep_clean)}

prod_id = np.asarray(prod_cat["ID"]).astype(str)

# a spectrum is usable if it is not entirely NaN (i.e. inside the VDFI footprint)
has_spec = np.array([np.isfinite(spec_all[i]).any() for i in range(n_prod)])

agn_idx, gal_idx = [], []
for i in range(n_prod):
    if not has_spec[i]:
        continue
    pid = prod_id[i]
    if agn_of.get(pid, False):
        agn_idx.append(i)
    elif clean_of.get(pid, False):
        gal_idx.append(i)

agn_idx = np.array(agn_idx, int); gal_idx = np.array(gal_idx, int)
print(f"product rows: {n_prod}   with real spectra: {int(has_spec.sum())}")
print(f"  AGN: {len(agn_idx)}    clean galaxies: {len(gal_idx)}    "
      f"(NaN/footprint-skipped already removed)")
```

## 2. The cycler widget

◀ / ▶ step through the current category; the dropdown jumps to any galaxy; the radio
buttons switch between Galaxy and AGN lists.

```python
z_add = 0
# ============================================================
#  CONFIG  (edit these two blocks if needed)
# ============================================================
# Rest-frame UV lines (vacuum Å). kind controls color/emphasis:
#   lya = Lyα, agn = AGN-sensitive (N V, C IV, He II, Si IV, C III]),
#   ism = interstellar absorption, neb = nebular.
UV_LINES = [
    ("Lyβ",       1025.72, "ism"),
    ("O VI",      1033.82, "agn"),
    ("Lyα",       1215.67, "lya"),
    ("N V",       1240.14, "agn"),
    ("Si II",     1260.42, "ism"),
    ("O I+Si II", 1303.00, "ism"),
    ("C II",      1334.53, "ism"),
    ("Si IV",     1397.76, "agn"),
    ("Si II",     1526.71, "ism"),
    ("C IV",      1549.48, "agn"),
    ("He II",     1640.42, "agn"),
    ("O III]",    1663.48, "neb"),
    ("Al II",     1670.79, "ism"),
    ("C III]",    1908.73, "agn"),
]
LINE_COLORS = {"lya": "red", "agn": "#1f77b4", "ism": "0.55", "neb": "#2ca02c"}

# AGN arms -> candidate column names in the catalog row. First match wins.
# >>> If a title says "per-arm columns not found", fix the names here. <
AGN_ARMS = [
    ("X-ray (Chandra)", ["AGN_XRAY", "AGN_X", "XRAY_AGN", "IS_XRAY_AGN", "AGN_XRAY_FLAG"]),
    ("Optical BPT",     ["AGN_OPT", "AGN_BPT", "BPT_AGN", "AGN_OPTICAL", "AGN_OPT_FLAG"]),
    ("IR (Donley)",     ["AGN_IR", "AGN_IRAC", "DONLEY_AGN", "AGN_IR_FLAG"]),
    ("MEx (Juneau)",    ["AGN_MEX", "MEX_AGN", "AGN_MEX_FLAG"]),
]
AGN_OVERALL = ["AGN_FLAG", "IS_AGN", "AGN"]

# ============================================================
#  HELPERS
# ============================================================
# AGN arms: (label, flag column, usability/applied gate or None)
AGN_ARMS = [
    ("X-ray (Chandra)", "AGN_XRAY", None),
    ("Optical BPT",     "AGN_OPT",  "AGN_OPT_USABLE"),
    ("IR (Donley)",     "AGN_IR",   "AGN_IR_USABLE"),
    ("MEx (Juneau)",    "AGN_MEX",  "AGN_MEX_APPLIED"),  # mass-guarded high-z arm
]
AGN_OVERALL = ["AGN_FLAG", "IS_AGN", "AGN"]

def _truthy(v):
    """Robust True/False from bool / int / float / string / masked scalar."""
    if v is None:
        return False
    if hasattr(v, "mask"):
        try:
            if bool(np.atleast_1d(v.mask)[0]):
                return False
        except Exception:
            pass
    if isinstance(v, (bytes, str)):
        s = (v.decode() if isinstance(v, bytes) else v).strip().lower()
        return s in ("true", "t", "1", "1.0", "yes", "y")   # "false" -> False
    try:
        if isinstance(v, float) and np.isnan(v):
            return False
    except Exception:
        pass
    try:
        return bool(v)
    except Exception:
        return False

def _row_has(row, name):
    try:
        return name in row.colnames
    except AttributeError:
        try:
            _ = row[name]; return True
        except Exception:
            return False

def _row_bool(row, name):
    return _truthy(row[name]) if _row_has(row, name) else False

def agn_mask_summary(row):
    """Which AGN arm(s) flagged this source, gated on data usability."""
    fired = []
    for label, flag_col, gate_col in AGN_ARMS:
        if not _row_bool(row, flag_col):
            continue
        if gate_col is not None and _row_has(row, gate_col) and not _row_bool(row, gate_col):
            continue  # arm fired but its diagnostic wasn't usable/applied -> don't credit
        fired.append(label)
    if fired:
        return "AGN — flagged by: " + ", ".join(fired)
    if any(_row_bool(row, c) for c in AGN_OVERALL):
        return "AGN — flagged (AGN_FLAG set, but no per-arm column was True)"
    return "galaxy — passes all AGN cuts"

def draw_uv_lines(ax, z, wave_obs):
    """Vertical markers for rest-UV lines, placed at λ_rest*(1+z) in the obs frame."""
    xlo, xhi = float(np.nanmin(wave_obs)), float(np.nanmax(wave_obs))
    for i, (name, lam0, kind) in enumerate(UV_LINES):
        lam = lam0 * (1.0 + z)
        if not (xlo <= lam <= xhi):
            continue
        c = LINE_COLORS[kind]
        ax.axvline(lam, color=c, ls="--", lw=0.8,
                   alpha=0.9 if kind in ("lya", "agn") else 0.5)
        ax.text(lam, 0.99 if i % 2 == 0 else 0.86, name,
                rotation=90, va="top", ha="right", fontsize=7, color=c,
                transform=ax.get_xaxis_transform(), clip_on=True)

# ============================================================
#  WIDGET
# ============================================================
state = {"cat": "Galaxy", "pos": 0}
cat_toggle = widgets.RadioButtons(options=["Galaxy", "AGN"], value="Galaxy",
                                  description="", layout=widgets.Layout(width="auto"))
prev_btn = widgets.Button(description="◀ Prev", layout=widgets.Layout(width="90px"))
next_btn = widgets.Button(description="Next ▶", layout=widgets.Layout(width="90px"))
jump = widgets.Dropdown(description="Jump:", layout=widgets.Layout(width="260px"))
id_box = widgets.Text(description="Go to ID:", placeholder="e.g. 12615",
                       layout=widgets.Layout(width="220px"))
out = widgets.Output()

def cur_list():
    return agn_idx if state["cat"] == "AGN" else gal_idx

def refresh_jump():
    lst = cur_list()
    opts = []
    for k, gi in enumerate(lst):
        cs = core.select_core_by_index(prod, int(gi))
        opts.append((f"{k}: ID {cs.row['ID']}  z={float(cs.row['z']):.3f}", k))
    jump.options = opts
    if len(opts):
        jump.value = min(state["pos"], len(opts)-1)

def draw():
    lst = cur_list()
    with out:
        out.clear_output(wait=True)
        if len(lst) == 0:
            print("(no spectra in this category)"); return
        state["pos"] %= len(lst)
        gi = int(lst[state["pos"]])
        cs = core.select_core_by_index(prod, gi)
        z  = float(cs.row["z"]) + z_add
        wr, fr, er = core.to_rest_frame(cs.wave, cs.flux, cs.err, z=z)

        tag = "AGN" if state["cat"] == "AGN" else "galaxy"
        mask_str = agn_mask_summary(cs.row)
        print(f"[{tag} {state['pos']+1}/{len(lst)}]  row {gi}  "
                     f"ID {cs.row['ID']}  z={z:.4f}  field={cs.row['FIELD']}\n"
                     f"{mask_str}")
        
        fig, ax = plt.subplots(2, 1, figsize=(12, 8))

        # --- top: observed frame, with rest-UV line markers ---
        ax[0].step((cs.wave)[20:-20], (cs.flux)[20:-20], "k", lw=0.8)
        if cs.err is not None:
            ax[0].fill_between((cs.wave)[20:-20], (cs.flux)[20:-20]-(cs.err)[20:-20], (cs.flux)[20:-20]+(cs.err)[20:-20], alpha=0.2, lw=0)
        ax[0].axhline(0, color="0.6", lw=0.7)
        ax[0].set_xlim(np.nanmin(cs.wave), np.nanmax(cs.wave))
        draw_uv_lines(ax[0], z, cs.wave)
        ax[0].set_xlabel("observed wavelength [Å]"); ax[0].set_ylabel("flux")
        ax[0].set_title("observed frame  (rest-UV lines at λ·(1+z))")

        # --- bottom: rest frame near Lyα ---
        sel = (wr > LYA_REST-REST_HALF_WIN) & (wr < LYA_REST+REST_HALF_WIN)
        ax[1].step(wr[sel], fr[sel], "k", lw=0.8)
        if er is not None:
            ax[1].fill_between(wr[sel], (fr-er)[sel], (fr+er)[sel], alpha=0.2, lw=0)
        ax[1].axvline(LYA_REST, color="r", ls="--", lw=0.8)
        ax[1].axhline(0, color="k", lw=0.8)
        ax[1].set_xlabel("rest wavelength [Å]"); ax[1].set_title("rest frame near Lyα")
        plt.tight_layout(); plt.show()

def on_prev(_):
    state["pos"] -= 1; jump.value = state["pos"] % len(cur_list()); draw()
def on_next(_):
    state["pos"] += 1; jump.value = state["pos"] % len(cur_list()); draw()
def on_cat(ch):
    state["cat"] = ch["new"]; state["pos"] = 0; refresh_jump(); draw()
def on_jump(ch):
    if ch["new"] is not None:
        state["pos"] = int(ch["new"]); draw()

def _norm_id(x):
    """Normalize IDs so 12615, '12615', and 12615.0 all compare equal."""
    s = str(x).strip()
    try:
        f = float(s)
        if f.is_integer():
            return str(int(f))
    except (ValueError, TypeError):
        pass
    return s

def goto_id(target):
    target = _norm_id(target)
    if not target:
        return

    ids = np.asarray(prod["catalog"]["ID"]).astype(str)
    match_rows = [i for i, v in enumerate(ids) if _norm_id(v) == target]

    with out:
        if not match_rows:
            print(f"Go to ID: no source with ID {target!r} in the catalog.")
            return

    gi = int(match_rows[0])
    agn_set = {int(x) for x in agn_idx}
    gal_set = {int(x) for x in gal_idx}

    if gi in agn_set:
        cat_name, lst = "AGN", agn_idx
    elif gi in gal_set:
        cat_name, lst = "Galaxy", gal_idx
    else:
        with out:
            print(f"Go to ID {target}: row {gi} isn't in the Galaxy or AGN selection.")
        return

    pos = int(np.flatnonzero(np.asarray(lst, int) == gi)[0])

    if state["cat"] != cat_name:
        cat_toggle.value = cat_name    # fires on_cat -> refresh_jump + draw
    state["pos"] = pos
    refresh_jump()
    jump.value = pos                   # sync dropdown
    draw()                             # guarantee a redraw even if value unchanged

def on_id_submit(change):
    goto_id(change["new"])

prev_btn.on_click(on_prev); next_btn.on_click(on_next)
cat_toggle.observe(on_cat, names="value")
jump.observe(on_jump, names="value")
id_box = widgets.Text(description="Go to ID:", placeholder="e.g. 12615",
                      continuous_update=False,          # fire only on Enter/blur
                      layout=widgets.Layout(width="220px"))
id_box.observe(on_id_submit, names="value")             # replaces id_box.on_submit(...)

display(widgets.VBox([
    widgets.HBox([cat_toggle, prev_btn, next_btn, jump, id_box]),
    out,
]))
draw()
#AGN numbers
#0, 4, 24, 30, 32, 40, 46, 51, 52, 
#Incorrect z:
#15-111: 25890, 16-115: 17563
#        +0.103         +0.15
```
